```python
loaded_list = []
for line in tmp_list:
    ll = line.split("  ")[:-1]
    print(len(line))
    print(len(ll))
    curr_tweet = list()
    for i in range(0, len(ll), 3):
        curr_tweet.append([ll[i], ll[i+1], ll[i+2]])
    loaded_list.append(curr_tweet)
```
