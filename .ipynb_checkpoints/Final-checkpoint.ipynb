# Custom tokenizer

```python
emoticons_str = r"""
    (?:
        [:=;] # Eyes
        [oO\-]? # Nose (optional)
        [D\)\]\(\]/\\OpP] # Mouth
    )"""
 
regex_str = [
    emoticons_str,
    r'<[^>]+>', # HTML tags
    r'(?:@[\w_]+)', # @-mentions
    r"(?:\#+[\w_]+[\w\'_\-]*[\w_]+)", # hash-tags
    r'http[s]?://(?:[a-z]|[0-9]|[$-_@.&amp;+]|[!*\(\),]|(?:%[0-9a-f][0-9a-f]))+', # URLs
 
    r'(?:(?:\d+,?)+(?:\.?\d+)?)', # numbers
    r"(?:[a-z][a-z'\-_]+[a-z])", # words with - and '
    r'(?:[\w_]+)', # other words
    r'(?:\S)' # anything else
]
    
tokens_re = re.compile(r'('+'|'.join(regex_str)+')', re.VERBOSE | re.IGNORECASE)
emoticon_re = re.compile(r'^'+emoticons_str+'$', re.VERBOSE | re.IGNORECASE)
    
def tokenize(s):
    return tokens_re.findall(s)
```

# Preprocessor with stemming, lemmatization or nothing

```python
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wn.ADJ
    elif treebank_tag.startswith('V'):
        return wn.VERB
    elif treebank_tag.startswith('N'):
        return wn.NOUN
    elif treebank_tag.startswith('R'):
        return wn.ADV
    else:
        return None 

def preprocess(s, lowercase=False, tokenizer=tokenize, word_transformation = '', handle_negation = True):
    tokens = tokenizer(s)
    if word_transformation == 'stem':
        stemmer = PorterStemmer()
        tokens = [stemmer.stem(token) for token in tokens]
    elif word_transformation == 'lemmatize':
        lemmatizer = WordNetLemmatizer()
        tagged = pos_tag(tokens)
        tokens = []
        for word, tag in tagged:
            wntag = get_wordnet_pos(tag)
            if wntag is None:
                lemma = lemmatizer.lemmatize(word) 
            else:
                lemma = lemmatizer.lemmatize(word, pos=wntag)
            tokens.append(lemma)        
    if lowercase:
        tokens = [token if emoticon_re.search(token) else token.lower() for token in tokens]
    if handle_negation:
        tokens = mark_negation(tokens)
    return tokens
t = time.time()
for tw in tweetsA:
    preprocess(tw, lowercase=True, word_transformation='')
tm = time.time()-t
print("done in {}".format(tm))
    
```

```python
class SentiWordNet_Extractor(BaseEstimator, TransformerMixin):

    def __init__(self):
        self.columns = ["SentiWordNet_pos_pos", "SentiWordNet_neg_pos", "SentiWordNet_pos_neg", "SentiWordNet_neg_neg", "SentiWordNet_neu_neg", "SentiWordNet_neu_pos"] 

    def transform(self, doc, y=None):
        pos_positives, pos_negatives, neg_negatives, neg_positives, neu_negatives, neu_positives = [], [], [], [], [], []
        lines = [preprocess(line, word_transformation='lemmatize') for line in doc]
        for line in tqdm(lines):
            pos_pos, pos_neg, neg_neg, neg_pos, neu_neg, neu_pos = 0, 0, 0, 0, 0, 0
            tagged = pos_tag(line)
            for word, tag in tagged:
                wntag = get_wordnet_pos(tag)
                synsets = wn.synsets(word.strip('_NEG'), pos=wntag)
                if not synsets:
                    continue
                synset = synsets[0]
                swn_synset = swn.senti_synset(synset.name())
                pos = swn_synset.pos_score()
                neu = swn_synset.obj_score()
                neg = swn_synset.neg_score()
#                 if word.endswith("_NEG"):
#                     pos_neg += pos
#                     neg_neg += neg
#                     neu_neg += neu
#                 else:
#                     pos_pos += pos
#                     neg_pos += neg
#                     neu_pos += neu
#                 print([pos, neu, neg])
                index = np.argmax([pos, neu, neg])
                if index == 0:
                    if word.endswith("_NEG"):
                        pos_neg += 1
                    else:
                        pos_pos += 1
                elif index == 1:
                    if neg >= pos:
                        if word.endswith("_NEG"):
                            neu_neg += 1
                            neg_neg += 1
                        else:
                            neu_pos += 1
                            neg_pos += 1
                    else:
                        if word.endswith("_NEG"):
                            neu_neg += 1
                            pos_neg += 1
                        else:
                            neu_pos += 1
                            pos_pos += 1
                else:
                    if word.endswith("_NEG"):
                        neg_neg += 1
                    else:
                        neg_pos += 1

            pos_positives.append(pos_pos)
            neg_positives.append(neg_pos)
            pos_negatives.append(pos_neg)
            neg_negatives.append(neg_neg)
            neu_negatives.append(neu_neg)
            neu_positives.append(neu_pos)
            
        pos_positives = np.array(pos_positives)[:,np.newaxis]
        neg_positives = np.array(neg_positives)[:,np.newaxis]
        pos_negatives = np.array(pos_negatives)[:,np.newaxis]
        neg_negatives = np.array(neg_negatives)[:,np.newaxis]
        neu_negatives = np.array(neu_negatives)[:,np.newaxis]
        neu_positives = np.array(neu_positives)[:,np.newaxis]
        return csr_matrix(np.hstack((pos_positives, neu_positives, neg_positives, pos_negatives, neu_negatives, neg_negatives,)), dtype=np.float64)
    

    def transformdf(self, doc):
        features = pd.SparseDataFrame(self.transform(doc), columns=self.columns)
        return features
    
    def fit(self, X, y=None):
        return self  # generally does nothing
print(SentiWordNet_Extractor().transform(tweetsA[:10]).toarray())
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from tqdm import trange, tqdm
from scipy import io
from itertools import groupby 
from nltk.corpus import opinion_lexicon
import preprocessor as p


class BingLiu_Extractor(BaseEstimator, TransformerMixin):
    
    
    def __init__(self):
        self.opinion_lexicon = opinion_lexicon
        self.columns = ["Bing_Liu_pos_pos", "Bing_Liu_neg_pos", "Bing_Liu_pos_neg", "Bing_Liu_neg_neg"]
    def transform(self, doc, y=None):
        pos_positives, neg_positives, pos_negatives, neg_negatives = [], [], [], []        
        olp = set(self.opinion_lexicon.positive())
        oln = set(self.opinion_lexicon.negative())
        lines = [preprocess(line, word_transformation='lemmatize', lowercase=True) for line in doc]
        for line in tqdm(lines):
            pos_pos, pos_neg, neg_neg, neg_pos = 0, 0, 0, 0
            for word in line:
                if word.endswith("_NEG"):
                    w = word.strip('_NEG')
                    if w in olp:
                        pos_neg +=1
                    elif w in oln:
                        neg_neg +=1
                else:
                    if word in olp:
                        pos_pos +=1
                    elif word in oln:
                        neg_pos +=1
            pos_positives.append(pos_pos)
            neg_positives.append(neg_pos)
            pos_negatives.append(pos_neg)
            neg_negatives.append(neg_neg)
        pos_positives = np.array(pos_positives)[:,np.newaxis]
        neg_positives = np.array(neg_positives)[:,np.newaxis]
        pos_negatives = np.array(pos_negatives)[:,np.newaxis]
        neg_negatives = np.array(neg_negatives)[:,np.newaxis]
        csr = csr_matrix(np.hstack((pos_positives,neg_positives,pos_negatives,neg_negatives)))
        io.mmwrite('bingliu_feature.mtx',csr)
        return csr
    
    def transformdf(self, doc):
        features = pd.SparseDataFrame(self.transform(doc), columns=self.columns)
        return features
    
    def fit(self, X, y=None):
        return self 

print(BingLiu_Extractor().transform(tweetsA[:10]).toarray())
```

```python
class Sentiment140_Extractor(BaseEstimator, TransformerMixin):

    def __init__(self):
        self.columns = ["Sentiment140_pos_pos", "Sentiment140_neg_pos", "Sentiment140_pos_neg", "Sentiment140_neg_neg"]  # e.g. pass in a column name to extract
        self.file = open("./lexicons/unigrams-pmilexicon.txt", "r")
        self.words = {}
        for line in self.file:
            parts = line.strip("\n").split("\t")
            word = parts[0]
            scores = [float(x) for x in parts[1:]]
            self.words[word] = scores
            
    def transform(self, doc, y=None):
        pos_positives, pos_negatives, neg_negatives, neg_positives = [], [], [], []
        lines = [preprocess(line, word_transformation='', lowercase=True) for line in doc]
        for line in tqdm(lines):
            pos_pos, pos_neg, neg_neg, neg_pos = 0, 0, 0, 0
            for word in line:
                if word.endswith("_NEG"):
                    w = word.strip('_NEG')
                    if w in self.words:
                        score = self.words[w][0]
                        if score > 0:
                            pos_neg += 1
                        else:
                            neg_neg += 1
                elif word in self.words:
                    score = self.words[word][0]
                    if score >0:
                        pos_pos += 1
                    else:
                        neg_pos += 1
            pos_positives.append(pos_pos)
            neg_positives.append(neg_pos)
            pos_negatives.append(pos_neg)
            neg_negatives.append(neg_neg)
            
        pos_positives = np.array(pos_positives)[:,np.newaxis]
        neg_positives = np.array(neg_positives)[:,np.newaxis]
        pos_negatives = np.array(pos_negatives)[:,np.newaxis]
        neg_negatives = np.array(neg_negatives)[:,np.newaxis]
        return csr_matrix(np.hstack((pos_positives, neg_positives, pos_negatives, neg_negatives)), dtype=np.float64)
    
    def transformdf(self, doc):
        features = pd.SparseDataFrame(self.transform(doc), columns=self.columns)
        return features
    
    def fit(self, X, y=None):
        return self  # generally does nothing
print(Sentiment140_Extractor().transform(tweetsA[:10]).toarray())
```

```python
class Custom_Extractor(BaseEstimator, TransformerMixin):

    def __init__(self):
        self.mycompile = lambda pat:  re.compile(pat,  re.UNICODE)  # e.g. pass in a column name to extract
        self.NormalEyes = r'[:=]'
        self.NoseArea = r'(|o|O|-)'   ## rather tight precision, \S might be reasonable...
        self.HappyMouths = r'[D\)\]]'
        self.SadMouths = r'[\(\[]'
        self.Happy_RE =  self.mycompile( '(\^_\^|' + self.NormalEyes + self.NoseArea + self.HappyMouths + ')')
        self.Sad_RE = self.mycompile(self.NormalEyes + self.NoseArea + self.SadMouths)
        self.Dotdotdot_RE = self.mycompile(r"\.\s?\.\s?\.")
        self.Mention_RE = self.mycompile(r'(?:@[\w_]+)') # @-mentions
        self.Hashtag_RE = self.mycompile(r"(?:\#+[\w_]+[\w\'_\-]*[\w_]+)") # hash-tags
        self.URL_RE = self.mycompile(r'http[s]?://(?:[a-z]|[0-9]|[$-_@.&amp;+]|[!*\(\),]|(?:%[0-9a-f][0-9a-f]))+') # URLs
        self.Elongated_RE = self.mycompile(r"(.)\1{2}")
        self.columns = ['# exclamations', '# interrogations', '# positive_emoticons', '# negative_emoticons',
                        '# uppercases', '# dotdotdot', '# quotations', '# mentions', '# hashtags', '# urls', '# elongated', '# negated']

    def transform(self, doc, y=None):
        exclamations = []
        interrogations = []
        positive_emoticons = []
        negative_emoticons = []
        uppercases = []
        dotdotdot = []
        quotations = []
        mentions = []
        hashtags = []
        urls = []
        elongated = []
        negated = []


        for line in doc:
            count = Counter(c if c in ['!','?','"'] else 'uppercase' if c.isupper() else None for c in line)
            exclamations.append(count['!'])
            interrogations.append(count['?'])
            quotations.append(count['"']/2)
            uppercases.append(count['uppercase'])
            count = Counter('negated' if c.endswith('_NEG') else None for c in preprocess(line))
            negated.append(count['negated'])
            positive_emoticons.append(len(self.Happy_RE.findall(line)))
            negative_emoticons.append(len(self.Sad_RE.findall(line)))
            dotdotdot.append(len(self.Dotdotdot_RE.findall(line)))
            mentions.append(len(self.Mention_RE.findall(line)))
            hashtags.append(len(self.Hashtag_RE.findall(line)))
            urls.append(len(self.URL_RE.findall(line)))
            elongated.append(len(self.Elongated_RE.findall(line)))
        return csr_matrix(np.hstack((np.array(exclamations)[:,np.newaxis], np.array(interrogations)[:,np.newaxis], 
                                     np.array(positive_emoticons)[:,np.newaxis], np.array(negative_emoticons)[:,np.newaxis], 
                                     np.array(uppercases)[:,np.newaxis], np.array(dotdotdot)[:,np.newaxis],
                                     np.array(quotations)[:,np.newaxis], np.array(mentions)[:,np.newaxis], 
                                     np.array(hashtags)[:,np.newaxis], np.array(urls)[:,np.newaxis], 
                                     np.array(elongated)[:,np.newaxis], np.array(negated)[:,np.newaxis]
                                   )))
        
    def transformdf(self, doc):
        features = pd.SparseDataFrame(self.transform(doc), columns=self.columns)
        return features

    def fit(self, X, y=None):
        return self  # generally does nothing


print(Custom_Extractor().transform(tweetsA[:10]).toarray())
```

```python
class Word2VecProvider(object):
    word2vec = None
    dimensions = 0

    def load(self, path_to_word2vec):
        self.word2vec = gensim.models.KeyedVectors.load_word2vec_format(path_to_word2vec, binary=False)
        self.word2vec.init_sims(replace=True)
        self.dimensions = self.word2vec.vector_size

    def get_vector(self, word):
        if word not in self.word2vec.vocab:
            return None

        return self.word2vec.syn0norm[self.word2vec.vocab[word].index]

    def get_similarity(self, word1, word2):
        if word1 not in self.word2vec.vocab or word2 not in self.word2vec.vocab:
            return None

        return self.word2vec.similarity(word1, word2)
```

```python
class word2vecExtractor(BaseEstimator, TransformerMixin):

    def __init__(self):
        self.word2vec_provider = word2vec
#         self.word2vec_provider = Word2VecProvider()
#         self.word2vec_provider.load(".\glove.twitter.27B.200d.txt")
        
    def transform(self, doc, y=None):
        word2vec_features = []
        for line in doc:
            similarities_good = []
            similarities_bad = []
            similarities_information = []
            current_word2vec = []
            current_similarities = []
            tokens = preprocess(line)
            for word in tokens:
                vec = self.word2vec_provider.get_vector(word.lower())
                if vec is not None:
                    current_word2vec.append(vec)
                sim_good = self.word2vec_provider.get_similarity(word.lower(), 'good')
                sim_bad = self.word2vec_provider.get_similarity(word.lower(), 'bad')
                sim_information = self.word2vec_provider.get_similarity(word.lower(), 'information')
                if sim_good != None :
                    similarities_good.append(sim_good)
                if sim_bad != None :
                    similarities_bad.append(sim_bad)
                if sim_information != None :
                    similarities_information.append(sim_information)
            similarities_good = np.mean(similarities_good)
            similarities_bad = np.mean(similarities_bad)
            similarities_information = np.mean(similarities_information)
            averaged_word2vec = list(np.array(current_word2vec).mean(axis=0))
            averaged_word2vec.append(similarities_good)
            averaged_word2vec.append(similarities_bad)
            averaged_word2vec.append(similarities_information)
            word2vec_features.append(averaged_word2vec)


        return np.array(word2vec_features)
        

    def fit(self, X, y=None):
        return self  # generally does nothing
```
